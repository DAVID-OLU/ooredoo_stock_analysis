```python
data = data.sort_values(by='Date')
```

### Calculating Daily Returns

### to show how much the stock price changes from one day to the next (in %)

```python
data['Daily Return'] = data['Adj Close'].pct_change()
```

### Calculating Moving Average

#### 7-day moving average:

```python
data['MA7'] = data['Adj Close'].rolling(window=7).mean()
```

#### 30-day movng average

```python
data['MA30'] = data['Adj Close'].rolling(window=30).mean()
```

### Volatility (Risk Measurement)

```python
#### high value -> price is unstable
#### Low value -> stock is stable
```

```python
data['Volatility'] = data['Daily Return'].rolling(window=7).std()
```

### Volume Trend
#### To check when trading activity increases

```python
data['Volume_MA7'] = data['Volume'].rolling(window=7).mean()
```

```python
data = data.dropna()
```
